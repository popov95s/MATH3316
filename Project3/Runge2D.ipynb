# Runge2D testing script

Loading all data output by Runge_uniform.cpp and Runge_Chebyshev.cpp

```python
%pylab inline
p6_uni = loadtxt('p6_uni.txt')
p24_uni = loadtxt('p24_uni.txt')
p6_Cheb = loadtxt('p6_Cheb.txt')
p24_Cheb = loadtxt('p24_Cheb.txt')
Runge = loadtxt('Runge.txt')
a= loadtxt('avals.txt')
b= loadtxt('bvals.txt')
```

Now plot f(x,y) first

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig = figure()
ax = fig.add_subplot(111, projection='3d')
X, Y = meshgrid(b, a)
surf = ax.plot_surface(X, Y, Runge, rstride=1, cstride=1, linewidth=0, cmap=cm.jet)
ax.set_xlabel('x')
ax.set_ylabel('y')
title('$f(x,y)$')
```

Plot the error $|f(x,y) - p_6|$ for uniformly-spaced nodes by iterating over both matrices and getting the corresponding difference

```python
error6_uni=numpy.ndarray((201,101))
for i in range(0,201):
    for j in range(0,101):
      error6_uni[i][j]= abs(Runge[i][j]- p6_uni[i][j])
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig = figure()
ax = fig.add_subplot(111, projection='3d')
X, Y = meshgrid(b, a)
surf = ax.plot_surface(X, Y, error6_uni, rstride=1, cstride=1, linewidth=0, cmap=cm.jet)
ax.set_xlabel('x')
ax.set_ylabel('y')
title('error $|f(x,y)-p_{6uni}|$')
```
